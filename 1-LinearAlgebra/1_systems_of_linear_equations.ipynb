# Linear Algebra and Machine Learning
## Systems of linear equations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

### Linear Regression
#### Supervised Machine Learning

```python
df = pd.read_csv('../0-Data/generatedpower_windspeed.csv', header=0, sep=',')
```

```python
model = LinearRegression().fit(df[['wind_speed(m/s)']], df[['power_output(kW)']])
M = model.coef_[0][0]
b = model.intercept_[0]
print(f"y = {M:.2f}*x + ({b:.2f})")
```

```python
df['predicted_power_(kW)'] = df['wind_speed(m/s)'].apply(lambda x:M*x + b)
```

```python
fig, ax = plt.subplots()
df.plot.scatter(x='wind_speed(m/s)', y='power_output(kW)', marker='o', c='black', ax=ax)
df.plot.line(x='wind_speed(m/s)', y='predicted_power_(kW)', ax=ax)
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Power Output (kW)')
plt.grid(linestyle=':')
plt.xlim((0,12))
plt.ylim((0,3800))
plt.legend(['Power Output (kW)', 'Linear Regression'])
plt.show()
```
